```python
import pandas as pd
data_pev = pd.read_csv('DATASET_PEV.csv')
data_av = pd.read_csv('DATASET_APRES_VENTE.csv')

# Nettoyage des colonnes inutiles
data_pev = data_pev.drop(columns=['Unnamed: 0'])
data_av = data_av.drop(columns=['Unnamed: 0'])

print(data_pev)
print(data_av)
```

```python
data = data_pev.groupby('VIN')['CALCULATEUR'].agg(' + '.join).reset_index()
df_merged_CALCULATEUR = pd.merge(data, data_av, on='VIN', how='inner')

df_merged_CALCULATEUR
```
